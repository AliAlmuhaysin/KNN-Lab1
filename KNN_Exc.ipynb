```python
def FindDistance(x1, y1):
    x2, y2 = 0 , 18
    point1 = np.array((x1,y1))
    point2 = np.array((x2,y2))
    return np.linalg.norm(point1 - point2)
```

```python
dist = []
for i in range(0,10):
    x1 = int(Data['Gender'][i])
    y1 = int(Data['Age'][i])
    dist.append(FindDistance(x1,y1))
```
